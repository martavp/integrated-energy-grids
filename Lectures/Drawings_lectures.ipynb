```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2 

x = np.linspace(0, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.savefig('fig1.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return x**4-(7*x)**2+2000 

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([-10,10])
plt.ylim([0,6000])
plt.savefig('fig2.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2
x = np.linspace(0, 10, 100)
plt.plot(x, f(x), color='blue')


colors=['gold', 'darkorange', 'red']
for i,k in enumerate([2, 0.5, 0.1]):
    def f(x):
       return (x-5)**2 -k*np.log(x-6)
    plt.plot(x, f(x), color=colors[i], label=k)
    
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.legend(title='k')
plt.savefig('fig4.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True
V=0.8
R=2
def f(x):
   return V*np.cos(x) 

def g(x):
   return (V/R)*np.cos(x) 
x = np.linspace(0, 10, 100)

plt.plot(x, f(x), color='blue', linewidth=3)
plt.plot(x, g(x), color='orange', linewidth=3)
plt.grid()
plt.xlabel('t')
plt.ylabel('')
plt.xlim([0,10])
plt.xticks([])
plt.ylim([-1,1])
plt.yticks([])
plt.savefig('fig6.png')
```

```python
plt.rcParams["figure.figsize"] = [4, 3]
plt.rcParams["figure.autolayout"] = True
r=0.0001
def f(x):
   return r*x**2 

x = np.linspace(-600, 600, 100)

plt.plot(x, f(x), color='blue')
plt.fill_between(x, f(x), 8.5, where=(f(x) <8.5), color='blue', alpha=0.6)

plt.grid()
plt.xlabel('$p_l$ (MW)')
plt.ylabel('$\Psi_l$ (MW)')
plt.ylim([-2,15])
plt.xlim([-600,600])
plt.savefig('fig10.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return np.cos(x) 

def g(x):
   return np.sin(x)
x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='red', linewidth=3, label='cos')
plt.plot(x, g(x), color='green', linewidth=3, label='sin')
plt.grid()
plt.ylabel('')
plt.xlim([-2,2])
plt.ylim([-1,1])
plt.yticks([-1,0,1])
plt.legend()
plt.savefig('fig11.png')
```
